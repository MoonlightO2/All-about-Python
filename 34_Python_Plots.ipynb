### **1. Line Plot:**

### **2. Scatter Plot:**

### **3. Histogram:**

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

### **4. Bar Plot:**

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 45]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

### **5. Stacked Bar Plot:**

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values1 = [25, 40, 30, 35]
values2 = [15, 30, 25, 40]

plt.bar(categories, values1, label='Group 1')
plt.bar(categories, values2, bottom=values1, label='Group 2')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Plot')
plt.legend()
plt.show()
```

### **6. Grouped Bar Plot:**

```python
import numpy as np
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values1 = [25, 40, 30, 35]
values2 = [15, 30, 25, 40]

x = np.arange(len(categories))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, values1, width, label='Group 1')
rects2 = ax.bar(x + width/2, values2, width, label='Group 2')

ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Grouped Bar Plot')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

plt.show()
```

### **13. Radar Chart (Spider Plot):**

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['A', 'B', 'C', 'D', 'E']
stats = [20, 30, 40, 50, 60]

angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()

stats += stats[:1]
angles += angles[:1]

plt.polar(angles, stats, marker='o')
plt.fill(angles, stats, 'blue', alpha=0.1)

plt.show()
```

### **26. Candlestick Chart:**

```python
pip install mplfinance
```

```python
import matplotlib.pyplot as plt
from mplfinance.original_flavor import candlestick_ohlc
import matplotlib.dates as mdates
import pandas as pd
import numpy as np

# Load the Iris dataset (not used in the candlestick chart)
from sklearn.datasets import load_iris
iris = load_iris()

# Generate random data for demonstration
np.random.seed(0)
num_points = 20
dates = pd.date_range(start='2024-01-01', periods=num_points)
open_prices = np.random.rand(num_points) * 10 + 50
high_prices = open_prices + np.random.rand(num_points) * 5
low_prices = open_prices - np.random.rand(num_points) * 5
close_prices = (high_prices + low_prices) / 2

# Create a DataFrame for candlestick chart
data = pd.DataFrame({'Date': dates, 'Open': open_prices, 'High': high_prices, 'Low': low_prices, 'Close': close_prices})
data['Date'] = data['Date'].map(mdates.date2num)

# Plot the candlestick chart
fig, ax = plt.subplots()
candlestick_ohlc(ax, data.values, width=0.6, colorup='g', colordown='r', alpha=1.0)
ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)
plt.title('Candlestick Chart - Random Data')
plt.xlabel('Date')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

### **31. lm Plot:**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
iris_df = sns.load_dataset("iris")

# Create an lm plot
sns.lmplot(x="sepal_length", y="sepal_width", data=iris_df, hue="species", markers=["o", "s", "D"])

# Set the title and labels
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Show the plot
plt.show()
```

### **32. Swarm Plot:**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
iris_df = sns.load_dataset("iris")

# Create a swarm plot
sns.swarmplot(x="species", y="petal_length", data=iris_df)

# Set the title and labels
plt.title('Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')

# Show the plot
plt.show()
```

### **34. Rel Plot:**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Auto MPG dataset
auto_mpg = sns.load_dataset("mpg")

# Create a relational plot (relplot)
sns.relplot(x="horsepower", y="mpg", hue="origin", size="weight", sizes=(40, 400), alpha=0.7, palette="muted", height=6, data=auto_mpg)

# Set the title and labels
plt.title('MPG vs Horsepower')
plt.xlabel('Horsepower')
plt.ylabel('Miles Per Gallon (MPG)')

# Show the plot
plt.show()
```
